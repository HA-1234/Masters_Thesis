Filtering on the basis of RE library

```python
trans_df = read_excel("back translated after manual annotation.xlsx")
trans_df
```

```python
trans_df = trans_df.drop(['English_Sentences', 'Normalized Sentences'], axis=1)
```

```python
import re

regex = re.compile('[^a-zA-Z]')

def similar_sentences(one, two):
    return regex.sub('', one) == regex.sub('', two)

def list_of_similar_sentences(one, two):
    return [similar_sentences(x, y) for (x, y) in zip(one, two)]
```

```python
# create new column which indicates whether original sentence == backtranslated sentence
trans_df['similar'] = list_of_similar_sentences(list(trans_df["Sentences"]), list(trans_df["Back_to_German_Sentences"]))
trans_df
```

```python
#create new df with only the sentences that have different backtranslation
df_with_diff_comp_of_translation = trans_df[trans_df["similar"]==False]
df_with_diff_comp_of_translation
```

```python
# new df with just different back translated sentences
df_with_diff_translation = df_with_diff_comp_of_translation[['Back_to_German_Sentences','Topics']]
df_with_diff_translation
```

```python
df_with_diff_translation = df_with_diff_translation.rename(columns={"Back_to_German_Sentences": "Sentences"})
df_with_diff_translation
```

```python
df_original = trans_df[['Sentences','Topics']]
df_original
```

```python
# Combining both df
both_dfs = [df_original, df_with_diff_translation]
result = pd.concat(both_dfs)
result
```
