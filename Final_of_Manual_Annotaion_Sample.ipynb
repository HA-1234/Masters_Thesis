```python
import pandas as pd
from pandas import read_excel
```

```python
df = read_excel("Data for Manual Annotation.xlsx")
df
```

```python
n1 = 1
n2 = 1000
filtered_df = pd.DataFrame({'Sentences' : [], 'English_Sentences': [], 'Topic labels 1': [], 'Topic labels 2': []})

while(n2<=6500):


  sampled_data = df[n1:n2].sample(n=40)
  filtered_df = pd.concat([filtered_df, sampled_data])

  if (n2 == 6000):
    sampled_data = df[n2:].sample(n=10)
    filtered_df = pd.concat([filtered_df, sampled_data])

  n1 += 1000
  n2 += 1000

filtered_df
```

```python
filtered_df.to_excel("Data for Co-annotation all columns.xlsx")
```
